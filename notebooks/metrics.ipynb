### Import Libraries

```python
import json
import matplotlib.pyplot as plt
```

### Import Metrics

```python
metrics = {}
optimizers = ['adam','adamw','rmsprop']

for opt in optimizers:
    with open(f'../models/{opt}_metrics.json', 'r') as file:
        data = json.load(file)
        metrics[opt] = {
            'loss': data[0],
            'acc': data[1],
            'training_time': data[2]
        }
```

### Loss

```python
plt.figure(figsize=(8, 6))

epoch_range = range(1, 21)

plt.plot(epoch_range, metrics['adam']['loss'], label='Adam', color='r')
plt.plot(epoch_range, metrics['adamw']['loss'], label='AdamW', color='g')
plt.plot(epoch_range, metrics['rmsprop']['loss'], label='RMSProp', color='b')

plt.xlabel('Epoch')
plt.xticks(epoch_range)
plt.ylabel('Loss')
plt.title('Testing Loss')

plt.legend()

plt.show()
```

### Accuracy

```python
plt.figure(figsize=(8, 6))

epoch_range = range(1, 21)

plt.plot(epoch_range, metrics['adam']['acc'], label='Adam', color='r')
plt.plot(epoch_range, metrics['adamw']['acc'], label='AdamW', color='g')
plt.plot(epoch_range, metrics['rmsprop']['acc'], label='RMSProp', color='b')

plt.xlabel('Epoch')
plt.xticks(epoch_range)
plt.ylabel('Accuracy (%)')
plt.title('Testing Accuracy')

plt.legend()

plt.show()
```
